```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("studentscores.csv")
```

```python
# Drop Unnamed column
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
plt.figure(figsize= (5,6))
ax = sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])

plt.show()
```

```python
# from the above chart we have analyzed that: 
# the number of female children is more than male children
```

```python
gd = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gd)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gd, annot = True)
plt.title("Relation b/w Parent's Education and Student's Scores")
plt.show()
```

```python
# from the above heatmap we can say that:
# the parent's education status impacts the children's overall scores
```

```python
gd1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gd1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gd1, annot = True)
plt.title("Relation b/w Parent's Matiral Status and Student's Scores")
plt.show()
```

```python
# from the above heatmap we can say that:
# the parent's marital status has no/negligible impact on the student's overall scores
```

```python
gd2 = df.groupby("WklyStudyHours").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gd2)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gd2, annot = True)
plt.title("Relation b/w Weekly Study Hours and Student's Scores")
plt.show()
```

```python
# from the above heatmap we can say that:
# the weekly study hours impact the student's overall score in Maths
```

```python
# Distribution of Ethnic Groups
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

glist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

label = ["groupA","groupB","groupC","groupD","groupE"]
plt.pie(glist, labels = label, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
